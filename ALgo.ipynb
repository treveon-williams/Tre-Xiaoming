```python
# Download data for ishares U.S. Healthcare ETF 'IYH'
IYH_df = yf.download("IYH",
                 start="2018-03-01",
                 end="2023-03-01",
                 progress=False)
display(IYH_df.head())
display(IYH_df.tail())
```

```python
# Download data for ishares U.S. Healthcare ETF 'IYF'
IYF_df = yf.download("IYF",
                 start="2018-03-01",
                 end="2023-03-01",
                 progress=False)
display(IYF_df.head())
display(IYF_df.tail())
```
